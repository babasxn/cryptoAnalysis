```python
# Data Preparation
def prepare_data(data,n=20, print_first_2_samples = True, no_of_features=1): # n--> no. of features

  # Create a list to store the values of training samples and the predicted samole
  X_train = []
  y_train = []

  data = data.values.reshape((-1,1))

  # Iterate over the data and store the n-1th samples in the train list and the nth sample in the predicted list
  for i in range(n, len(data)):
    X_train.append(data[i-20:i, 0]) #--- i-60 -- i-1
    y_train.append(data[i,0])

    # Print the first 2 samples
    if i<=21 and print_first_2_samples == True:
      print(X_train)
      print()
      print(y_train)
  
  # Convert X_train and y_train into numpy arrays
  X_train = np.array(X_train)
  y_train = np.array(y_train)

  # Reshape the data into [rows, time_steps, features]
  X_train_updated = X_train.reshape((X_train.shape[0], X_train.shape[1], no_of_features)) # no_of_features=1 for univariate time series

  return X_train_updated, y_train
```

```python
# 1. Deep RNN Model

# importing sequential model, simple rnn and time distributed layer
from keras.models import Sequential 
from keras.layers import SimpleRNN #, LSTM, GRU
from keras.layers import TimeDistributed
from keras.layers import Dense

# Model object
rnn_model = Sequential()

# adding SimpleRNN layers
rnn_model.add(SimpleRNN(50, activation='tanh', return_sequences=True, input_shape = [None, 1]))
#rnn_model.add(LSTM(50, activation='tanh', return_sequences=True, input_shape = [None, 1]))
#rnn_model.add(GRU(50, activation='tanh', return_sequences=True, input_shape = [None, 1]))
#--> Dense
rnn_model.add(SimpleRNN(50, activation='tanh', return_sequences=True))
#--> Dense
rnn_model.add(SimpleRNN(50, activation='tanh', return_sequences=True))
#--> Dense
rnn_model.add(SimpleRNN(50, activation='tanh', return_sequences=True))
#--> Dense
rnn_model.add(SimpleRNN(50, activation='tanh', return_sequences=True))
#--> Dense
rnn_model.add(SimpleRNN(50, activation='tanh', return_sequences=True))
#--> Dense
rnn_model.add(SimpleRNN(50, activation='tanh', return_sequences=True))
#--> Dense

# add the Timedistributed layer
rnn_model.add(TimeDistributed(Dense(1, activation='relu'))) 

# Printing the summary
rnn_model.summary()
```

```python
# Cmpute the metrics
from keras.metrics import mean_squared_error as MSE
def last_time_step_mse(y_true, y_pred):
  return MSE(y_true[ : , -1], y_pred[ : , -1])

# Compile the model
rnn_model.compile(optimizer='adam', loss='mse', metrics=[last_time_step_mse])
```
